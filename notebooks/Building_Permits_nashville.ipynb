2) Make 2 tables from this file the first one as Residential contractors and the 2 table as Commercial 

```python
# Create the Residential contractors DataFrame by filtering based on 'Permit Type Description'
residential_contractors = Building_permits_nashville[
    Building_permits_nashville['Permit Type Description'].str.contains('Residential', case=False, na=False)
]
```

```python
# Create the Commercial contractors DataFrame by filtering based on 'Permit Type Description'
commercial_contractors = Building_permits_nashville[
    Building_permits_nashville['Permit Type Description'].str.contains('Commercial', case=False, na=False)
]
```

```python
# Display the first few rows of each DataFrame for verification
print("Residential Contractors:")
print(residential_contractors.head())
```

```python
print("\nCommercial Contractors:")
print(commercial_contractors.head())
```

3) Construction Cost Column do from highest to lowest value for each new table

```python
# Sort the Residential contractors DataFrame by 'Construction Cost' in descending order
residential_contractors_sorted = residential_contractors.sort_values(by='Construction Cost', ascending=False)
```

```python
# Sort the Commercial contractors DataFrame by 'Construction Cost' in descending order
commercial_contractors_sorted = commercial_contractors.sort_values(by='Construction Cost', ascending=False)
```

```python
# Display the first few rows of each sorted DataFrame for verification
print("Residential Contractors Sorted by Construction Cost:")
print(residential_contractors_sorted.head())
```

```python
print("\nCommercial Contractors Sorted by Construction Cost:")
print(commercial_contractors_sorted.head())
```

```python
# Save the Residential contractors sorted DataFrame to a CSV file
residential_contractors_sorted.to_csv('Residential_Contractors_Sorted.csv', index=False)
```

```python
# Save the Commercial contractors sorted DataFrame to a CSV file
commercial_contractors_sorted.to_csv('Commercial_Contractors_Sorted.csv', index=False)
```

```python
# Define the mapping dictionary for permit codes to descriptions
permit_type_mapping = {
    'CACA': 'Building Commercial Addition',
    'CACF': 'Building Commercial Foundation',
    'CACG': 'Building Commercial Structural Frame',
    'CACH': 'Building Commercial Shell',
    'CACJ': 'Building Commercial Fire Damage',
    'CACK': 'Building Commercial Roofing/Siding',
    'CACL': 'Building Commercial Rehab Storm Damage',
    'CACN': 'Building Commercial New',
    'CACR': 'Building Commercial Rehab',
    'CACS': 'Building Commercial New Storm Damage',
    'CACT': 'Building Commercial Tenant Finish Out',
    'CACP': 'Building Commercial Change of Contractor'
}



```

```python
# Create a new column 'Permit Description' based on the 'Permit Type' column
commercial_contractors['Permit Description'] = commercial_contractors['Permit Type'].map(permit_type_mapping)
```

```python
# Display the first few rows of the updated DataFrame for verification
print(commercial_contractors.head())
```

```python
commercial_contractors.to_csv('Commercial_Contractors.csv', index=False)
```

```python
# Define the permit types to filter
permit_types_to_include = ['CACA', 'CACN', 'CACR', 'CACT']
```

```python
# Filter the commercial_contractors DataFrame for the specified permit types
selected_permits = commercial_contractors[commercial_contractors['Permit Type'].isin(permit_types_to_include)]
```

```python
# Display the first few rows of the new filtered DataFrame for verification
print(selected_permits.head())
```

```python
selected_permits.to_csv('Selected_Commercial_Permits.csv', index=False)
```
